# Chairing a meeting
- There should be an agenda
- Get the agenda items preplanned and have everyone read it before the meeting (but everyone needs to do the pre-read)
- Everyone must be heard/have an opportunity to put their points forward, make sure no one is sidelined
- Keep the meeting productive and on time, don't let it drag too long
- Example meeting rules:
    - Ask people to speak ‘through the chair’. This means putting your hand up if you want to speak, and waiting for the chair to say it’s your turn.
    - Don’t interrupt other people.
    - Stick to the item on the agenda.
    - Don’t talk amongst yourselves.
    - Respect other people’s views – don’t groan or pull faces when someone else is speaking. Wait until they’ve finished and then put your point of view calmly and politely.
    - Keep contributions short and to the point.
    - Start and finish the meeting on time.

## Getting through the business on time
- Don't assume people know what you're talking about
- Stick to the discussion

## Involving everyone
- Restrain those who talk too much
    - If someone's talking too long, give them a time limit
    - Don't let people aggressively interrupt
- Encourage the quieter ones to speak up
    - Give priority to quieter ones to speak
    - Make sure minority/unpopular views are heard and considered
- Encourage collaboration

## Reaching decisions
- Keep an overview of the meeting, document key points
- Write down what was decided, a summary of what was said
    - There needs to be clear notes of the meeting so no one is unsure what next
- Get an agreement of what's decided, silence =/= agreement

## Dealing with difficult people
- Keep some rules agreed on beforehand
- Be firm and consistent, don't let people abuse the rules
- Keep the discussion on the points/facts, not people
- If someone won't listen, get a collective vote when ready to move on

# Virtual meetings
- Don't let people go on mute and leave for a bit
- Ban multitasking
    - Get people to share their thoughts
    - Give people different tasks to do
- Conclude with having everyone go around and say what they would've done differently (speak now or forever hold your peace)

## Common rules
- Be respectful of other's time
- Be present
