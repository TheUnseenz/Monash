**g)Why is the cross-entropy loss caused by the feed-forward neural network at $(x,y)$ (i.e., $\text{CE}(1_y, p(x))$) always non-negative? When does this $\text{CE}(1_y, p(x))$ loss get the value $0$? Note that you need to answer this question for a general pair $(x,y)$ and a general feed-forward neural network with, for example $M=4$ classes?**  
* **Why is it always non-negative?**  
    The loss simplifies to $-\log(p(x)_c)$, where $p(x)_c$ is the predicted probability for the correct class $c$. Since $p(x)_c$ is a probability, its value is in the range $[0, 1]$. The logarithm of a number in $(0, 1]$ is always non-positive ($\le 0$). Multiplying by -1 makes the final result non-negative ($\ge 0$).

* **When is the loss zero?**  
    The loss, $-\log(p(x)_c)$, equals 0 if and only if $\log(p(x)_c) = 0$. This occurs only when $p(x)_c = 1$. This means the loss is zero only when the model predicts the correct class with 100% confidence, which is a perfect prediction.

**(a)** You need to develop the class `MyLinear` with the following skeleton. You need to declare the weight matrix and bias of this linear layer.

<div style="text-align: right"><span style="color:red">[3 points]</span></div>

```python
class MyLinear(torch.nn.Module):
  def __init__(self, input_size, output_size):
    """
    input_size: the size of the input
    output_size: the size of the output
    """
    super().__init__()
    #Your code here
    self.W = torch.nn.Parameter(torch.empty(output_size, input_size))
    self.b = torch.nn.Parameter(torch.empty(output_size))
    
    # Kaiming uniform init
    torch.nn.init.kaiming_uniform_(self.W, a=5**0.5)
    if self.b is not None:
        fan_in, _ = torch.nn.init._calculate_fan_in_and_fan_out(self.W)
        bound = 1 / fan_in**0.5 if fan_in > 0 else 0
        torch.nn.init.uniform_(self.b, -bound, bound)

  #forward propagation
  def forward(self, x): #x is a mini-batch
    #Your code here
    # Forward pass = Matrix multiplication + bias add
    return x @ self.W.T + self.b
  
```

```python
class MyFFN(torch.nn.Module):
  def __init__(self, input_size, num_classes, hidden_sizes, act = torch.nn.ReLU()):
    """
    input_size: the size of the input
    num_classes: the number of classes
    act is the activation function
    hidden_sizes is the list of hidden sizes
    for example input_size = 3, hidden_sizes = [5, 7], num_classes = 4, and act = torch.nn.ReLU()
    means that we are building up a FFN with the confirguration
    (3 (Input) -> 5 (ReLU) -> 7 (ReLU) -> 4 (Output))
    """
    super(MyFFN, self).__init__()
    self.input_size = input_size
    self.num_classes = num_classes
    self.act = act
    self.hidden_sizes = hidden_sizes
    self.num_layers = len(hidden_sizes) + 1

  def create_FFN(self):
    """
    This function creates the feed-forward neural network
    We stack many MyLinear layers
    """
    hidden_sizes = [self.input_size] + self.hidden_sizes + [self.num_classes]
    self.layers = []
    #Your code here
    self.velocities = {}
    self.accelerations = {}
    for input_shape, output_shape in zip(hidden_sizes, hidden_sizes[1:]):
      self.layers.append(MyLinear(input_shape,output_shape))

  def forward(self,x):
    """
    This implements the forward propagation of the batch x
    This needs to return the logits of x
    """
    #Your code here
    output = x
    for i, layer in enumerate(self.layers):
      output = layer(output)
      if i < len(self.layers) - 1:  # no activation on last layer
          output = self.act(output)
    return output

  def compute_loss(self, x, y):
    """
    This function computes the cross-entropy loss
    You can use the built-in CE loss of PyTorch
    """
    #Your code here
    criterion = torch.nn.CrossEntropyLoss()
    return criterion(x, y)

  def update_SGD(self, x, y, learning_rate = 0.01):
    """
    This function updates the model parameters using SGD using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    # Zero gradients
    for layer in self.layers:
      for param in layer.parameters():
            if param.grad is not None:
                param.grad = None
    # Forward pass
    output = self.forward(x)
    # Compute loss
    loss = self.compute_loss(output, y)
    # Backward pass
    loss.backward()
    # Parameter update
    with torch.no_grad():  # prevent autograd tracking
      for layer in self.layers:
          for param in layer.parameters():
              if param.grad is None:
                  continue
              param -= learning_rate * param.grad

    

  def update_SGDwithMomentum(self, x, y, learning_rate = 0.01, momentum = 0.9):
    """
    This function updates the model parameters using SGD with momentum using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    # Zero gradients
    last_state = []
    for layer in self.layers:
      for param in layer.parameters():
            last_state.append(param.grad)
            if param.grad is not None:
                param.grad = None
    # Forward pass
    output = self.forward(x)
    # Compute loss
    loss = self.compute_loss(output, y)
    # Backward pass
    loss.backward()
    # Parameter update
    with torch.no_grad():  # prevent autograd tracking
      for layer in self.layers:
          for param in layer.parameters():
              if param.grad is None:
                  continue
              # Initialize velocity if first time
              if param not in self.velocities:
                  self.velocities[param] = torch.zeros_like(param)
              self.velocities[param] = momentum*self.velocities[param] + param.grad
              param -= learning_rate * self.velocities[param]  

  def update_AdaGrad(self, x, y, learning_rate = 0.01):
    """
    This function updates the model parameters using AdaGrad using the batch (x,y)
    You need to implement the update rule manually and cannot rely on the built-in optimizer
    """
    #Your code here
    eps = 1e-8
    # Zero gradients
    for layer in self.layers:
      for param in layer.parameters():
            if param.grad is not None:
                param.grad = None
    # Forward pass
    output = self.forward(x)
    # Compute loss
    loss = self.compute_loss(output, y)
    # Backward pass
    loss.backward()
    # Parameter update
    with torch.no_grad():  # prevent autograd tracking
      for layer in self.layers:
          for param in layer.parameters():
              if param.grad is None:
                  continue
              # Initialize velocity if first time
              if param not in self.accelerations:
                  self.accelerations[param] = torch.zeros_like(param)
              self.accelerations[param] = self.accelerations[param] + param.grad**2
              param -= learning_rate * param.grad/(self.accelerations[param]**0.5 + eps)
```

**(c)** Write the code to evaluate the accuracy of the current `myFFN` model on a data loader (e.g., train_loader or test_loader).

```python
def compute_acc(model, data_loader):
  """
  This function computes the accuracy of the model on a data loader
  """
  #Your code here  
  size = len(data_loader.dataset)
  num_batches = len(data_loader)
  model.eval()
  test_loss, correct = 0, 0

  with torch.no_grad():
      for X, y in data_loader:
          # X, y = X.to(device), y.to(device)
          pred = model(X)
          test_loss += model.compute_loss(pred, y).item()
          correct += (pred.argmax(1) == y).type(torch.float).sum().item()

  test_loss /= num_batches
  accuracy = correct / size
  # print(f"Avg loss: {test_loss:>8f}, Accuracy: {(100*accuracy):>0.1f}%")
  return accuracy
```

#### <span style="color:red">**Question 2.2:**</span> Write the code for the feed-forward neural net using PyTorch

<div style="text-align: right"> <span style="color:red">[5 points]</span> </div>

We now develop a feed-forward neural network with the architecture $784 \rightarrow 40(ReLU) \rightarrow 30(ReLU) \rightarrow 10(softmax)$. You can choose your own way to implement your network and an optimizer of interest. You should train model in $50$ epochs and evaluate the trained model on the test set.

```python
#Your code here
from torch import nn
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"Using {device} device")

class NeuralNetwork(nn.Module):
    act_functions = [nn.Sigmoid, nn.Tanh, nn.ReLU]
    def __init__(self, act=2, n1=40, n2=20):
        super().__init__()
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(784, n1),
            self.act_functions[act](),
            nn.Linear(n1, n2),
            self.act_functions[act](),
            nn.Linear(n2, 10) # No final softmax as cross-entropy loss function already includes a softmax in it.
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
    
model = NeuralNetwork().to(device)
print(model)

# Parameters
learning_rate = 0.0001
epochs = 50

# Initialize the loss function
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)


# Training loop
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)

        # Forward pass
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss_val, current = loss.item(), batch * len(X)
            print(f"loss: {loss_val:>7f}  [{current:>5d}/{size:>5d}]")


def evaluate(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    accuracy = correct / size
    print(f"Avg loss: {test_loss:>8f}, Accuracy: {(100*accuracy):>0.1f}%")
    return accuracy


# Run training for 50 epochs
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_loader, model, loss_fn, optimizer)
    evaluate(val_loader, model, loss_fn)

print("Training complete!")

# Final evaluation on the test set
print("Evaluating on test set...")
test_acc = evaluate(test_loader, model, loss_fn)
print(f"Final Test Accuracy: {(100*test_acc):.2f}%")
```

<div style="text-align: right"> <span style="color:red">[5 points]</span> </div>

Sharpness-aware minimization (SAM) (i.e., [link for main paper](https://openreview.net/pdf?id=6Tm1mposlrM) from Google Deepmind) is a simple yet but efficient technique to improve the generalization ability of deep learning models on unseen data examples. In your research or your work, you might potentially use this idea. Your task is to read the paper and implement *Sharpness-aware minimization (SAM)*. Finally, you need to apply SAM to the best architecture found in **Question 2.3**.   

```python
#Your code here
#https://github.com/davda54/sam
class SAM(torch.optim.Optimizer):
    def __init__(self, params, base_optimizer, rho=0.05, adaptive=False, **kwargs):
        assert rho >= 0.0, f"Invalid rho, should be non-negative: {rho}"

        defaults = dict(rho=rho, adaptive=adaptive, **kwargs)
        super(SAM, self).__init__(params, defaults)

        self.base_optimizer = base_optimizer(self.param_groups, **kwargs)
        self.param_groups = self.base_optimizer.param_groups
        self.defaults.update(self.base_optimizer.defaults)

    @torch.no_grad()
    def first_step(self, zero_grad=False):
        grad_norm = self._grad_norm()
        for group in self.param_groups:
            scale = group["rho"] / (grad_norm + 1e-12)

            for p in group["params"]:
                if p.grad is None: continue
                self.state[p]["old_p"] = p.data.clone()
                e_w = (torch.pow(p, 2) if group["adaptive"] else 1.0) * p.grad * scale.to(p)
                p.add_(e_w)  # climb to the local maximum "w + e(w)"

        if zero_grad: self.zero_grad()

    @torch.no_grad()
    def second_step(self, zero_grad=False):
        for group in self.param_groups:
            for p in group["params"]:
                if p.grad is None: continue
                p.data = self.state[p]["old_p"]  # get back to "w" from "w + e(w)"

        self.base_optimizer.step()  # do the actual "sharpness-aware" update

        if zero_grad: self.zero_grad()

    @torch.no_grad()
    def step(self, closure=None):
        assert closure is not None, "Sharpness Aware Minimization requires closure, but it was not provided"
        closure = torch.enable_grad()(closure)  # the closure should do a full forward-backward pass

        self.first_step(zero_grad=True)
        closure()
        self.second_step()

    def _grad_norm(self):
        shared_device = self.param_groups[0]["params"][0].device  # put everything on the same device, in case of model parallelism
        norm = torch.norm(
                    torch.stack([
                        ((torch.abs(p) if group["adaptive"] else 1.0) * p.grad).norm(p=2).to(shared_device)
                        for group in self.param_groups for p in group["params"]
                        if p.grad is not None
                    ]),
                    p=2
               )
        return norm

    def load_state_dict(self, state_dict):
        super().load_state_dict(state_dict)
        self.base_optimizer.param_groups = self.param_groups
    
model = NeuralNetwork(act=1, n1=40, n2= 40).to(device)
print(model)

# Parameters
learning_rate = 0.0001
epochs = 50

# Initialize the loss function
base_optimizer = torch.optim.Adam  # define an optimizer for the "sharpness-aware" update
optimizer = SAM(model.parameters(), base_optimizer, lr=learning_rate)


# Training loop
def train(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        X, y = X.to(device), y.to(device)

        def closure():
            optimizer.zero_grad()
            pred = model(X)
            loss = loss_fn(pred, y)
            loss.backward()
            return loss

        # First forward-backward step
        loss = closure()
        optimizer.first_step(zero_grad=True)

        # Second forward-backward step (recompute gradients at perturbed weights)
        closure()
        optimizer.second_step(zero_grad=True)

        if batch % 100 == 0:
            loss_val, current = loss.item(), batch * len(X)
            print(f"loss: {loss_val:>7f}  [{current:>5d}/{size:>5d}]")

def evaluate(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    model.eval()
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device), y.to(device)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    accuracy = correct / size
    print(f"Avg loss: {test_loss:>8f}, Accuracy: {(100*accuracy):>0.1f}%")
    return accuracy


# Run training for 50 epochs
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train(train_loader, model, loss_fn, optimizer)
    evaluate(val_loader, model, loss_fn)

print("Training complete!")

# Final evaluation on the test set
print("Evaluating on test set...")
test_acc = evaluate(test_loader, model, loss_fn)
print(f"Final Test Accuracy: {(100*test_acc):.2f}%")
```

```python
#Your code here
class YourBlock(nn.Module):
  def __init__(self, in_feature_maps, out_feature_maps, drop_rate = 0.2, batch_norm = True, use_skip = True):
    super(YourBlock, self).__init__()
    self.use_skip = use_skip
    #Your code here
    self.batch_norm = batch_norm
    layers = []
    layers.append(nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=3, stride=1, padding=1))
    if batch_norm:
        layers.append(nn.BatchNorm2d(out_feature_maps))
    layers.append(nn.ReLU(inplace=True))

    layers.append(nn.Conv2d(out_feature_maps, out_feature_maps, kernel_size=3, stride=1, padding=1))
    if batch_norm:
        layers.append(nn.BatchNorm2d(out_feature_maps))
    layers.append(nn.ReLU(inplace=True))

    layers.append(nn.MaxPool2d(kernel_size=2, stride=2))
    layers.append(nn.Dropout(p=drop_rate))

    self.block = nn.ModuleList(layers)

    # if skip connection and in/out mismatch, add a projection conv
    if use_skip:
        self.skip_conv = nn.Conv2d(in_feature_maps, out_feature_maps, kernel_size=1, stride=2)
    else:
        self.skip_conv = None


  def forward(self, x):
    #Write your code here
    identity = x
    for layer in self.block:
        x = layer(x)

    if self.use_skip:
        if self.skip_conv is not None:
            identity = self.skip_conv(identity)
        x = x + identity
    return x
```

```python
class YourCNN(nn.Module):
  def __init__(self, list_feature_maps = [16, 32, 64], drop_rate = 0.2, batch_norm= True, use_skip = True):
    super(YourCNN, self).__init__()
    layers = []
    #Write your code here
    # first conv + bn + relu
    layers.append(nn.Conv2d(3, list_feature_maps[0], kernel_size=3, stride=1, padding=1))
    if batch_norm:
        layers.append(nn.BatchNorm2d(list_feature_maps[0]))
    layers.append(nn.ReLU(inplace=True))

    # stack YourBlocks
    for i in range(1, len(list_feature_maps)):
        layers.append(
            YourBlock(
                list_feature_maps[i-1], 
                list_feature_maps[i], 
                drop_rate=drop_rate,
                batch_norm=batch_norm,
                use_skip=use_skip
            )
        )

    # global avg pool + flatten + linear classifier
    layers.append(nn.AdaptiveAvgPool2d((1, 1)))
    layers.append(nn.Flatten())
    layers.append(nn.Linear(list_feature_maps[-1], 20))

    self.block = nn.ModuleList(layers)



  def forward(self, x):
    #Write your code here
    for layer in self.block:
      x = layer(x)
    return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def compute_loss(model, loss_fn, loader):
  loss = 0
  # Set model to eval mode for inference
  model.eval()
  with torch.no_grad():  # No need to track gradients for validation
    for (batchX, batchY) in loader:
      # Move data to the same device as the model
      batchX, batchY = batchX.to(device).type(torch.float32), batchY.to(device).type(torch.long)
      loss += loss_fn(model(batchX), batchY)
  # Set model back to train mode
  model.train()
  return float(loss)/len(loader)
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def compute_acc(model, loader):
    correct = 0
    totals = 0
    # Set model to eval mode for inference
    model.eval()
    for (batchX, batchY) in loader:
        # Move batchX and batchY to the same device as the model
        batchX, batchY = batchX.to(device).type(torch.float32), batchY.to(device)
        outputs = model(batchX)  # feed batch to the model
        totals += batchY.size(0)  # accumulate totals with the current batch size
        predicted = torch.argmax(outputs.data, 1)  # get the predicted class
        # Move batchY to the same device as predicted for comparison
        correct += (predicted == batchY).sum().item()
    return correct / totals
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def fit(model= None, train_loader = None, valid_loader= None, optimizer = None,
        num_epochs = 50, verbose = True, seed= 1234):
  torch.manual_seed(seed)
  # Move the model to the device before initializing the optimizer
  model.to(device) # Move the model to the GPU

  if optimizer == None:
    optim = torch.optim.Adam(model.parameters(), lr = 0.001) # Now initialize optimizer with model on GPU
  else:
    optim = optimizer
  history = dict()
  history['val_loss'] = list()
  history['val_acc'] = list()
  history['train_loss'] = list()
  history['train_acc'] = list()

  for epoch in range(num_epochs):
    model.train()
    for (X, y) in train_loader:
      # Move input data to the same device as the model
      X,y = X.to(device), y.to(device)
      # Forward pass
      outputs = model(X.type(torch.float32)) # X is already on the correct device
      loss = loss_fn(outputs, y.type(torch.long))
      # Backward and optimize
      optim.zero_grad()
      loss.backward()
      optim.step()
      #losses and accuracies for epoch
    val_loss = compute_loss(model, loss_fn, valid_loader)
    val_acc = compute_acc(model, valid_loader)
    train_loss = compute_loss(model, loss_fn, train_loader)
    train_acc = compute_acc(model, train_loader)
    history['val_loss'].append(val_loss)
    history['val_acc'].append(val_acc)
    history['train_loss'].append(train_loss)
    history['train_acc'].append(train_acc)
    if verbose:
      print(f"Epoch {epoch+1}/{num_epochs}")
      print(f"train loss= {train_loss:.4f} - train acc= {train_acc*100:.2f}% - valid loss= {val_loss:.4f} - valid acc= {val_acc*100:.2f}%")
  return history
```

<div style="text-align: right"> <span style="color:red">[4 points]</span> </div>

Data mixup is another super-simple technique used to boost the generalization ability of deep learning models. You need to incoroporate data mixup technique to the above deep learning model and experiment its performance. There are some papers and documents for data mixup as follows:
- Main paper for data mixup [link for main paper](https://openreview.net/pdf?id=r1Ddp1-Rb) and a good article [article link](https://www.inference.vc/mixup-data-dependent-data-augmentation/).

You need to extend your model developed above, train a model using data mixup, and write your observations and comments about the result.

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

def mixed_augment(x, y, alpha=0.2, p_mixup=(1.0,0.0)):
    """Apply Mixup or CutMix with probability p_mixup."""
    device = x.device
    r = torch.rand(1, device=device).item()
    batch_size = x.size(0)
    index = torch.randperm(batch_size, device=device)

    if r < p_mixup[0]:
        # --- Mixup ---
        lam = torch.distributions.Beta(alpha, alpha).sample().to(device)
        mixed_x = lam * x + (1 - lam) * x[index]
        y_a, y_b = y, y[index]

    elif r < p_mixup[0] + p_mixup[1]:
        # --- CutMix ---
        lam = torch.distributions.Beta(alpha, alpha).sample().to(device)
        y_a, y_b = y, y[index]

        H, W = x.size(2), x.size(3)
        cut_rat = torch.sqrt(1. - lam)
        cut_w, cut_h = (W * cut_rat).int(), (H * cut_rat).int()

        cx = torch.randint(W, (1,), device=device)
        cy = torch.randint(H, (1,), device=device)

        bbx1 = torch.clamp(cx - cut_w // 2, 0, W)
        bby1 = torch.clamp(cy - cut_h // 2, 0, H)
        bbx2 = torch.clamp(cx + cut_w // 2, 0, W)
        bby2 = torch.clamp(cy + cut_h // 2, 0, H)

        mixed_x = x.clone()
        mixed_x[:, :, bby1:bby2, bbx1:bbx2] = x[index, :, bby1:bby2, bbx1:bbx2]

        lam = 1 - ((bbx2 - bbx1) * (bby2 - bby1) / (W * H))

    else:
        return x, y, y, 1.0

    return mixed_x, y_a, y_b, lam

def mix_criterion(criterion, pred, y_a, y_b, lam):
    return lam * criterion(pred, y_a) + (1 - lam) * criterion(pred, y_b)

def fit_mixed(
    model=None,
    train_loader=None,
    valid_loader=None,
    optimizer=None,
    num_epochs=20,
    verbose=True,
    seed=1234,
    alpha=0.2,
    p_mixup=(1.0,0.0),
):
    torch.manual_seed(seed)
    device = next(model.parameters()).device

    if optimizer is None:
        optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    history = {"train_loss": [], "train_acc": [], "val_loss": [], "val_acc": []}

    for epoch in range(num_epochs):
        # === Training phase ===
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for X, y in train_loader:
            X, y = X.to(device), y.to(device)

            # Mixup / CutMix
            mixed_x, y_a, y_b, lam = mixed_augment(X, y, alpha, p_mixup)

            outputs = model(mixed_x)
            loss = mix_criterion(loss_fn, outputs, y_a, y_b, lam)

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Track training loss
            running_loss += loss.item() * X.size(0)

            # Track training accuracy (use main label y_a for simplicity)
            _, preds = torch.max(outputs, 1)
            correct += (lam * preds.eq(y_a).sum() + (1 - lam) * preds.eq(y_b).sum()).item()
            total += X.size(0)

        epoch_train_loss = running_loss / total
        epoch_train_acc = correct / total

        # === Validation phase ===
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for X, y in valid_loader:
                X, y = X.to(device), y.to(device)
                outputs = model(X)
                loss = loss_fn(outputs, y)

                val_loss += loss.item() * X.size(0)
                _, preds = torch.max(outputs, 1)
                val_correct += preds.eq(y).sum().item()
                val_total += X.size(0)

        epoch_val_loss = val_loss / val_total
        epoch_val_acc = val_correct / val_total

        # Save history
        history["train_loss"].append(epoch_train_loss)
        history["train_acc"].append(epoch_train_acc)
        history["val_loss"].append(epoch_val_loss)
        history["val_acc"].append(epoch_val_acc)

        if verbose:
            print(
                f"Epoch {epoch+1}/{num_epochs} "
                f"- train loss={epoch_train_loss:.4f}, acc={epoch_train_acc*100:.2f}% "
                f"- val loss={epoch_val_loss:.4f}, acc={epoch_val_acc*100:.2f}%"
            )

    return history

# Optimal settings
best_lr = 0.001
best_use_skip = True

model = YourCNN(use_skip=best_use_skip).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=best_lr)
print("Begin training")
history = fit_mixed(
    model,
    train_loader,
    val_loader,
    optimizer=optimizer,
    num_epochs=20,
    verbose=True,
    p_mixup=(1.0,0.0)
)
```

<span style="color:red">**Question 3.5:**</span> Implement the **one-versus-all (OVA)** loss. The details are as follows:
- You need to apply `the sigmoid activation function` to logits $h = [h_1, h_2,...,h_M]$ instead of `the softmax activation` function as usual to obtain $p = [p_1, p_2,...,p_M]$, meaning that $p_i = sigmoid(h_i), i=1,...,M$. Note that $M$ is the number of classes.
- Given a data example $x$ with the ground-truth label $y$, the idea is to maximize the likelihood $p_y$ and to minimize the likelihoods $p_i, i \neq y$. Therefore, the objective function is to find the model parameters to
  - $\max\left\{ \log p_{y}+\sum_{i\neq y}\log(1-p_{i})\right\}$ or equivalently $\min\left\{ -\log p_{y}-\sum_{i\neq y}\log(1-p_{i})\right\}$.
  - For example, if $M=3$ and $y=2$, you need to minimize $\min\left\{ -\log(1-p_{1})-\log p_{2}-\log(1-p_{3})\right\}$.

Compare the model trained with the OVA loss and the same model trained with the standard cross-entropy loss.

<div style="text-align: right"> <span style="color:red">[4 points]</span> </div>



```python
#Your code here
class OVALoss(nn.Module):
    """
    One-vs-All (OVA) Loss.
    Equivalent to Binary Cross-Entropy over all classes with a one-hot target.
    """
    def __init__(self):
        super(OVALoss, self).__init__()

    def forward(self, logits, targets):
        """
        logits: [batch_size, num_classes] raw scores from model
        targets: [batch_size] integer class labels (0 <= y < num_classes)
        """
        # one-hot encode the targets
        num_classes = logits.size(1)
        target_onehot = F.one_hot(targets, num_classes=num_classes).float()

        # Apply BCE with logits: combines sigmoid + BCE in a numerically stable way
        loss = F.binary_cross_entropy_with_logits(logits, target_onehot, reduction="mean")

        return loss
    
loss_fn = OVALoss()

# Hyperparameter grids
p_mixup_grid = [
    (0.0, 1.0),
    (0.2, 0.8),
    (0.4, 0.6),
    (0.6, 0.4),
    (0.8, 0.2),
    (1.0, 0.0),
]
lr_grid = [0.001] # lr = 0.001 always best
alpha_grid = [0.1, 0.2, 0.3, 0.4]

# Ensure model save dir exists
save_dir = "./models"
os.makedirs(save_dir, exist_ok=True)

# Path to results JSON file
results_path = os.path.join(save_dir, "results.json")

# Load existing results if resuming
if os.path.exists(results_path):
    with open(results_path, "r") as f:
        results = json.load(f)
else:
    results = []


try:
    # Cartesian product of all combinations
    for (p_mixup, lr, alpha) in itertools.product(p_mixup_grid, lr_grid, alpha_grid):
        model_filename = f"model_ova_pm{p_mixup[0]}-{p_mixup[1]}_lr{lr}_alpha{alpha}.pt"
        model_path = os.path.join(save_dir, model_filename)

        if already_done(p_mixup, lr, alpha, "OVA"):
            print(f"Skipping {model_filename} (already in results.json).")
            continue

        print(f"\n=== Training with p_mixup={p_mixup}, lr={lr}, alpha={alpha} ===")

        # fresh model + optimizer
        model = YourCNN().to(device)
        optimizer = torch.optim.Adam(model.parameters(), lr=lr)

        # Train
        history = fit_mixed(
            model,
            train_loader,
            val_loader,
            optimizer=optimizer,
            num_epochs=50,
            p_mixup=p_mixup,
            alpha=alpha,
        )

        # Extract metrics
        best_train_acc = max(history["train_acc"]) if "train_acc" in history else None
        min_train_loss = min(history["train_loss"]) if "train_loss" in history else None
        best_val_acc = max(history["val_acc"]) if "val_acc" in history else None
        min_val_loss = min(history["val_loss"]) if "val_loss" in history else None

        # Save model checkpoint
        torch.save(model.state_dict(), model_path)

        # Save results entry
        entry = {
            "p_mixup": list(p_mixup),
            "lr": lr,
            "alpha": alpha,
            "train_acc": best_train_acc,
            "train_loss": min_train_loss,
            "val_acc": best_val_acc,
            "val_loss": min_val_loss,
            "model_path": model_path,
            "loss_type": "OVA"
        }
        results.append(entry)

        # Write to disk immediately (resume-safe)
        with open(results_path, "w") as f:
            json.dump(results, f, indent=2)

        # Print summary for this run
        print(f"✅ Completed: p_mixup={p_mixup}, lr={lr}, alpha={alpha} | "
              f"val_acc={best_val_acc:.4f}, val_loss={min_val_loss:.4f}, saved={model_path}")

except KeyboardInterrupt:
    print("\n⏹️ Interrupted by user. Saving progress...")
    with open(results_path, "w") as f:
        json.dump(results, f, indent=2)
    print("Progress saved. You can safely resume later.")

# Final summary sorted by val_acc
sorted_results = sorted(
    results,
    key=lambda x: (x["val_acc"] is not None, x["val_acc"]),
    reverse=True,
)

print("\n=== Final Results (sorted by val_acc) ===")
for r in sorted_results:
    print(f"p_mixup={tuple(r['p_mixup'])}, lr={r['lr']}, alpha={r['alpha']}, "
          f"val_acc={r['val_acc']:.4f}, val_loss={r['val_loss']:.4f}, "
          f"saved={r['model_path']}")
```

```python
#Your code here
import torch
import torch.nn.functional as F
import torch.optim as optim

class OVALoss(nn.Module):
    """
    One-vs-All (OVA) Loss.
    Equivalent to Binary Cross-Entropy over all classes with a one-hot target.
    """
    def __init__(self):
        super(OVALoss, self).__init__()

    def forward(self, logits, targets):
        """
        logits: [batch_size, num_classes] raw scores from model
        targets: [batch_size] integer class labels (0 <= y < num_classes)
        """
        # one-hot encode the targets
        num_classes = logits.size(1)
        target_onehot = F.one_hot(targets, num_classes=num_classes).float()

        # Apply BCE with logits: combines sigmoid + BCE in a numerically stable way
        loss = F.binary_cross_entropy_with_logits(logits, target_onehot, reduction="mean")

        return loss
    
loss_fn = OVALoss()
# === PGD attack function ===
def pgd_attack(model, X, y, eps=0.0313, eta=0.002, k=20, device="cuda"):
    model.eval()
    X_adv = X.clone().detach().to(device)
    X_adv = X_adv + 0.001 * torch.randn_like(X_adv).to(device)  # small random init
    X_adv.requires_grad = True

    for _ in range(k):
        outputs = model(X_adv)
        loss = loss_fn(outputs, y)   # swap with OVA loss if needed
        loss.backward()

        with torch.no_grad():
            X_adv = X_adv + eta * X_adv.grad.sign()
            delta = torch.clamp(X_adv - X, min=-eps, max=eps)
            X_adv = torch.clamp(X + delta, -1, 1)

        X_adv.requires_grad = True

    return X_adv.detach()

# === PGD adversarial training (clean + adversarial per batch) ===
def advTrainPGD(model, train_loader, num_epochs=25, learning_rate=0.001,
                eps=0.0313, eta=0.002, k=20, device="cuda"):
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    history = {"train_loss": [], "clean_acc": [], "adv_acc": []}
    best_robust_acc = 0.0
    for epoch in range(num_epochs):
        model.train()
        total, correct = 0, 0
        total_adv, correct_adv = 0, 0
        running_loss = 0.0

        for xs, ys in train_loader:
            xs, ys = xs.to(device), ys.to(device)

            # === 1. Clean training ===
            optimizer.zero_grad()
            preds = model(xs)
            loss = loss_fn(preds, ys)  # replace with OVA loss if needed
            loss.backward()
            optimizer.step()

            running_loss += loss.item() * xs.size(0)
            _, pred_labels = preds.max(1)
            correct += pred_labels.eq(ys).sum().item()
            total += ys.size(0)

            # === 2. Adversarial training ===
            xs_adv = pgd_attack(model, xs, ys, eps, eta, k, device)
            optimizer.zero_grad()
            preds_adv = model(xs_adv)
            loss_adv = loss_fn(preds_adv, ys)
            loss_adv.backward()
            optimizer.step()

            _, pred_labels_adv = preds_adv.max(1)
            correct_adv += pred_labels_adv.eq(ys).sum().item()
            total_adv += ys.size(0)

        clean_acc = correct / total
        adv_acc = correct_adv / total_adv
        epoch_loss = running_loss / total

        history["train_loss"].append(epoch_loss)
        history["clean_acc"].append(clean_acc)
        history["adv_acc"].append(adv_acc)

        print(f"Epoch [{epoch+1}/{num_epochs}] "
              f"Loss={epoch_loss:.4f}, Clean Acc={clean_acc*100:.2f}%, Adv Acc={adv_acc*100:.2f}%")
        
        # Save best model
        if adv_acc > best_robust_acc:
            best_robust_acc = adv_acc
            torch.save(model.state_dict(), robust_model_path)
            print(f"Best robust model saved (robust acc={adv_acc*100:.2f}%)")

    return model, history



# === Main script ===
best_model_path = "./models/best_model.pt"
robust_model_path = "./models/model_ova_pgd_trained.pt"

# Load base model
model = YourCNN().to(device)
model.load_state_dict(torch.load(best_model_path))

print("Starting adversarial training...")
model, history = advTrainPGD(model, train_loader, num_epochs=50,
                          eps=0.0313, eta=0.002, k=20, device=device)


# === Evaluate robust accuracy on test set ===
def evaluate_robust_accuracy(model, test_loader, eps=0.0313, eta=0.002, k=20, device="cuda"):
    model.eval()
    total, correct = 0, 0
    for X, y in test_loader:
        X, y = X.to(device), y.to(device)
        X_adv = pgd_attack(model, X, y, eps, eta, k, device)
        with torch.no_grad():
            outputs = model(X_adv)
            _, preds = outputs.max(1)
            correct += preds.eq(y).sum().item()
            total += y.size(0)
    return correct / total

clean_acc = compute_acc(model, val_loader)
robust_acc = evaluate_robust_accuracy(model, val_loader, eps=0.0313, eta=0.002, k=20, device=device)

print(f"\nFinal Evaluation:")
print(f"Clean Test Accuracy:  {clean_acc*100:.2f}%")
print(f"Robust Test Accuracy: {robust_acc*100:.2f}%")

```
