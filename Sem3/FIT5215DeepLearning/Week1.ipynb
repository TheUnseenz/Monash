# Recap

"Label" is best defined as values, categories or classes assigned to data points  
Probability of prediction = $\frac{e^x}{\sum_{X=x}{e^X}}$  
Cross entropy loss, for one-hot encoding simplifies to: -log $\frac{e^x}{\sum_{X=x}{e^X}}$  
Therefore, cross-entropy loss becomes 0 at p(ground truth) = 1. Range of cross entropy loss is 0 to positive infinity.    
For a discrete distribution, outputs must sum to 1
