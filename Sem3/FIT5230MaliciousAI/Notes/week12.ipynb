## Practical Realities of AI Governance  
Trust is your #1 currency; the earlier you can build it,, the better (also: horizontal vs vertical networking)  
Horizontal (fellow colleagues) often more helpful than vertical (seniors/juniors)  
Create win-win situations  
Treat everyone as a stakeholder  
There's a lot of uncertainty to be handled, mistakes will be made, everyone would make mistakes.  
Things will heat up; don't get angry with people doing their roles  
At best, everyone will be equally unhappy  
Make sure you get your contact available to the right people (journalists here)  
There's communicated red lines and there's actual red lines  
Optics matter, make presentation good
