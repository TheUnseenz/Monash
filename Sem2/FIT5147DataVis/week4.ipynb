### Spatial Data & Thematic Mapping
Spatial data is data that is naturally associated with a geographic location or region and for which
you want to explore the data taking this association into account.  

Data mapping is a data science term related to how you link elements of the data to elements of a
system, often to allow it to be moved or used.  

A data map (or thematic map as they are usually called by cartographers) is a visualisation used if
the data is associated with a geographic location or region and you wish to explore or show how this
affects the other attributes. Just because data has a geographic location doesn’t necessarily mean
that you should show it on a map.  

Common data maps:  
- Choropleth map  
    - Areas shaded or patterned according to data value  
- Prism map  
    - Height represents value  
- Flow map  
    - Directions of flow  
- Dot density map  
    - Dots per x value  
- Isarithmic map  
    - Contours connecting regions of similar values  
- Proportional symbol map  
    - Drawn symbols according to value  
- Cartogram  
    - Geographic size distorted according to data value

### Map Projections

A map projection is the mapping between points on the surface of
the Earth and the position on the two-dimensional map.  

There is no single best projection: the appropriate choice depends
upon area covered by the map and the purpose of the map.  

Projections can be classified depending upon which property they
preserve.  
A. Equal-area (or equivalent): the relative size of area is preserved  
B. Conformal: angles are preserved locally  
C. Equidistant: distances along some directions are preserved  
D. Compromise: neither area nor angles are preserved, but overall not “too bad”  
No projection can preserve areas and angles.  


Mercator projection is a poor choice for world maps, but often a
good default projection for maps of small areas.

### Developable Surfaces
- When selecting a projection for a map showing a section of Earth, there are 3 classical types of map projections
- they are grouped by their developable surface.  
- a developable surface is a geometric shape that can be unwrapped to a flat plane without introducing distortion.  
- three developable surfaces: cylinders, cones and planes/azimuthal.
