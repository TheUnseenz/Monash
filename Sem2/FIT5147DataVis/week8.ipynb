# Week 8 - Visualization Design Methodology

## 5 Design Sheet Methodology (5dS or FdS)
5 sheets  
5 parts per sheet  
5 parts to the process   

Sheet 1 = Brainstorming ideas
Sheet 2, 3, 4 = Main design 
Sheet 5 = Final polish

### Sheet 1 - Brainstorming  
Recall exercise of sketching potential ideas. Brainstorm possibilities, filter, categorize, combine andrefine ideas.
- Generate ideas:  
  - Short concepts that could be part of whole  
- Filter the ideas  
  - Remove duplicate ideas  
- Categorize the sketches  
  - Order the ideas  
- Combine & Refine  
  - Organise the ideas  
- Question  
  - Do this ideas provide a solution?

### Sheet 2, 3, 4 Initial Designs  
Sheets 2-4 are same sections, but COMPLETELY DIFFERENT **FULL** designs for visualization layout
- Layout  
- Focus/Zoom (key techniques)  
- Operations  
- Discussion  
  - Pros  
  - Cons  
- Meta-info  
  - Title, authors, date, sheet #, tasks  


### Sheet 5 - Realization Design  
Full design to be implemented. Can be either based on one of sheets 2, 3 or 4 OR
a combination of ideas from 2 or 3 sheets.  
It must NOT be a completely new design, and must NOT be all of sheets 2, 3 and 4.  
- Discussion -> Details
  - Description of Algorithms or Methods
  - Dependencies & Requirements
  - Estimates of Cost



## Workshop
Data Graphic Design Summary - Revisited (Week 1)  
Edward Tuftes principles:  
- Data Density: The data density is the proportion of the
  overall graph that is dedicated to displaying data.  
- Graphical Integrity: Present the data itself and avoid
  embedding misleading messages.  
- Data Correspondence: Help the viewer to understand the
  patterns and form hypotheses about the relationships in
  the data  
- Aesthetics: Data graphics should be aesthetically
  pleasing, but be mindful to keep it simple  
- Avoid chart junk
- Maximize data-ink ratio
