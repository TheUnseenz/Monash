## Network Data Analysis and Visualization

Network diameter: shortest path between any 2 nodes in network  
Networks are connected if there is a path between them  
Path is directed if edges in it are directed and flow from start to end of path  
 
Network Degree: number of edges incident to node (number of neighbours)  
Network Centrality:  
- Higher degree of n -> more important  
- Node is central if it is close to many other nodes  
- Betweenness centrality: the extent to which node n lies between other nodes  
- Status or rank centrality: a node is more important if it has important neighbours.

### Visual components
Reading direction  
    - Arrange things top left to bottom right, in order of things you want the reader to read  
Visual centre  
    - Slightly above geometric middle. Put important stuff here.  
Sight lines  
    - Align objects with each other to minimize sight lines  
Symmetry and Balance  
    - Symmetric and asymmetric can both look nice  
    - But don't put everything all on one side  
White space  
    - Space between visual elements  
    - Important to keep space between elements to partition elements  
Typography  
    - Sans serif are easier to read than serif. Use this for paragraphs. Clean and modern.  
    - Serif is good for title. Traditional.   

Readability of Text  
- Row length shouldn't be too long - keep within 60 characters  
- Font size - keep about 11-12 for digital, 9-10 for print  
- Encode with size/weight/italic/uppercase and mixed case/colour hue/combinations  
- Our eyes prioritize high contrast text  



## Workshop
### Networks
Networks encode relations (edges) between items (nodes).  
Both edges and nodes can have attributes.  

Key concepts  
• Neighbour  
• Directed  
• Shortest path  
• Connected  
• Degree: number of neighbours  
• Closeness: average distance  
• Betweenness: shortest paths  
• Status or rank: neighbour importance  
• Edge centrality  

https://datavizcatalogue.com/  

When to use:  
Arc Diagram  
  
- Good for showing correlation in ordinal data (MUST be ordered)  
- E.g.  
    - Sequence analysis i.e. DNA/RNA, musical structure, text analysis  
    - Temporal data/event sequences  
    - Network analysis  

Adjacency matrix  
  
- Good for showing relationships between nodes, spotting clusters
- Asymmetric matrix shows direction of relationships

Sunburst diagram  
  
- Shows hierarchy breakdown  
- Multiple rings to indicate levels of hierarchy  

Sankey diagram  
  
- Shows hierarchical relationships/breakdown  

Chord diagram  
  
- Show asymmetric relationship  

Network Earth Videographic  
- Arc diagram  
- Parallel coordinates plot/sankey/alluvial  
- Non-ribbon chord diagram  
- Network diagram  

Uses colour hue, value, saturation, size, orientation, position, shape to encode data  
They show interaction of the elements vs each other
