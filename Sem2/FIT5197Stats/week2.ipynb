## Week 2 notes - Probability Distribution Theory

We describe our data via probability models  
Recorded values are treated as realizations of random variables  
- This is due to experimental/measurement error  
- We cannot possibly record every confounding variable  

### Probability rules:
1. A probability distribution sums to 1  
2. The probability of A₁ OR A₂ happening is P(A₁ ∪ A₂) = P(A₁) + P(A₂) - P(A₁ ∩ A₂)  
where ∪ represents the union (OR) and ∩ represents the intersection (AND) of the events.  

Marginal probability = $P(X = x)$ regardless of value of other factors  

Conditional probability $P(X = x | Y = y) = \frac{P(X = x, Y = y)}{P(Y = y)}$  

or in words, $P(X)$ given observation of $Y=y$ is $P(X=x \text{ and } Y=y)$ divided by $P(Y=y)$   
<br>  
Bayes' Theorem:  
The probability of event x given event A is equal to the probability of event A given event x, multiplied by the probability of event x, all divided by the probability of event A.  
Bayes' theorem tells you how to update your beliefs about the probability of an event based on new evidence.  

Mathematically:  
$P(x|A) = \frac{P(A|x) P(x)}{P(A)}$  

This can also be expressed as:  

$P(x|A) = \frac{P(A|x) P(x)}{\sum_{x} P(A|x) P(x)}$  

Complement rule P(not A) = 1 - P(A)  

Product rule P(B∩A) = P(B|A) p(A)  

Sum rule P(A) = $\sum_{x}P(x ∩ A)$  

<br>
Probability density function = continuous version (as opposed to discrete). Integrates to 1, instead of summing to 1.  
Inverse CDF, quantile function finds x value where P(X <= x) is p  
E.g. Q(p = 0.5) is the median (an X value, not a probability or y value!)

P(only A) = P(A) - P(A n B) - P(A n C) + P(A n B n C)

Example:
Draw 5 cards from a deck of 52, with no jokers and no replacements
P(draw 5 cards, no reds) = (26!/21!)/(52!/47!) ~= 0.025
P(at least 1 ace from 5 cards) = 1 - P(no ace in 5 draws)
    = 1 - (48!/43!)/(52!/47!)
    ~= 0.34

Bayes Theorem:
$P(x|A) = \frac{P(A|x)P(x)}{P(A)}$
  
Derivation:
P(A|x) = P(A n x)/P(x)

$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$
Example:
P(A) = 0.01 has disease
P(B|A) = 0.9 tests positive
P(B|not A) = 0.15
P(B) = sum(P(B,A)) 
     = P(B|A)P(A) + P(B|not A)P(not A)
     = 0.01\*0.9 + 0.15\*0.99
     = 0.1575
    
P(A|B) = 0.9*0.01/0.1575
        = 0.057

number of ways selecting right or left
