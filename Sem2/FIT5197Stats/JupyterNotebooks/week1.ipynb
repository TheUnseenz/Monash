Sample standard deviation Sx = Square root of variance

```r
# Example:
x <- c(9.2, 14.1, 9.8, 12.4, 16.0, 12.6, 22.7, 18.9, 21.0, 14.5, 20.4, 16.9)
sum_x <- sum(x)
mean_x <- sum_x / length(x)
print(mean_x)

variance_x <- sum(((x - mean_x)^2)) / (length(x) - 1)
std_dev_x <- sqrt(variance_x)
print(std_dev_x)
```
