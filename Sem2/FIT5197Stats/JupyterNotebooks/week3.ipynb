P(only A) = P(A) - P(A n B) - P(A n C) + P(A n B n C)

Example:
Draw 5 cards from a deck of 52, with no jokers and no replacements
P(draw 5 cards, no reds) = (26!/21!)/(52!/47!) ~= 0.025
P(at least 1 ace from 5 cards) = 1 - P(no ace in 5 draws)
    = 1 - (48!/43!)/(52!/47!)
    ~= 0.34

Bayes Theorem:
$P(x|A) = \frac{P(A|x)P(x)}{P(A)}$
  
Derivation:
P(A|x) = P(A n x)/P(x)

$P(A|B) = \frac{P(B|A)P(A)}{P(B)}$
Example:
P(A) = 0.01 has disease
P(B|A) = 0.9 tests positive
P(B|not A) = 0.15
P(B) = sum(P(B,A)) 
     = P(B|A)P(A) + P(B|not A)P(not A)
     = 0.01\*0.9 + 0.15\*0.99
     = 0.1575
    
P(A|B) = 0.9*0.01/0.1575
        = 0.057

Expected Value E(x) = $\displaystyle\sum_{x\subset X} x.P(x)$
For continuous probability function, integrate instead of sum

Discrete probability functions are written as P(x), upper case  
Continuous probability functions are written as p(x), lower case  

Reminder example:  
simple triangular distribution p(x) = 1 - |x|  

$E[x] = \int_{-1}^{0} x(1+x) \,dx + \int_{0}^{1} x(1-x) \,dx $

$E[x] = \int_{1}^{0} a(1-a) \,da + \int_{0}^{1} x(1-x) \,dx $  
$E[x] = -\int_{0}^{1} a(1-a) \,da + \int_{0}^{1} x(1-x) \,dx $ = 0
