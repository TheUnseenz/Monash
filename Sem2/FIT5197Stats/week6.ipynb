### Central Limit Theorem
Data we have is not perfect. It *will* be messy and irregular  
Using CLT we can still analyze and make predictions  

To use CLT:  
1. Take *many* random samples from the data (>30)  
2. Calculate the mean of each sample  
3. Plot the means  

Results:  
- A normal (bell-shaped) distribution, even if the original distribution is not normal  

Why?  
- Law of large numbers  
- Means of random samples is essentially a convolution, which looks similar to normal distribution  
- Variance shrinks with larger samples (its divided by sample size)  

E.g.   
"How long do customers wait on average to get their order?"  
But 
- you can't track every single customer  
- wait times aren't constant  
What to do?  
1. Take random sample - 30 customers on Monday  
2. Calculate the average waiting time  
3. Repeat 1. and 2. many times across different days  
4. Plot the average waiting time  

#### Statement of Central Limit Theorem
Let S = Y1 + Y2... with E[Yi] = μ and Var[Yi] = σ^2  
For large n, distribution of S is approximately normal distribution with mean nμ  and variance nσ^2  
Sample sum, S = Y1+Y2... + Yn  
= Summation of Yi  
E[S] = Summation of E[Yn]

### Confidence Intervals
It is a way of expressing how confident we are that a certain parameter of interest falls within a specific range.  
Interval Estimation  
- Narrow interval -> lower uncertainty  
- Wide interval  

Common to consider: α = 0.05 (95% confidence interval)  

How to generate confidence interval?  
Given unknown mean, known variance,  
mean ~= N(mean, variance/n)  

Calculate estimate of mean = summation of Yi/n  
Estimate of mean - 1.96*σ/sqrt(n), Estimate of mean _ 1.96\*σ/sqrt(n)  
-> 95% confident that population mean is between the range above  

Given both mean and variance unknown  
No longer a standard normal distribution, but instead a student t distribution with t table instead of z table  
This includes (n-1) as an extra parameter, for degree of freedom  
Two tails = sum up both ends of the distribution
