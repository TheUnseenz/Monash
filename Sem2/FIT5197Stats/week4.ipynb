```python
sample_weights = [66, 54, 69, 77, 55, 97, 79, 52, 56, 60, 64, 102]
mean = sum(sample_weights)/len(sample_weights)
print(mean)
variance = 0
for item in sample_weights:
    variance += (item - mean)**2
variance = variance/(len(sample_weights)-1)
print(variance)
stddev = variance**0.5
left_range = mean-36
right_range = 100-mean
print(left_range, right_range, stddev)

# Transform normal pdf to standard normal pdf 
# for W = 36
Z_lower = (36-mean)/stddev
Z_upper = (100-mean)/stddev
print(Z_lower, Z_upper)
# above and below are the same in magnitude!
print(left_range/stddev, right_range/stddev)
# P(within 36 to 100) = P(100) - P(36)
# Dig up stats table for standard normal for answer
```

#### Binomial distribution
PMF of binomial distribution is:  
$P(m|Î¸) = \binom{n}{m}Î¸^m(1-Î¸)^{n-m}$

1. How should you interpret $(1-Î¸)^{n-m}$?  
Probability of seeing (n-m) failures  

2. How should you interpret $Î¸^m$?  
Probability of seeing m successes  

3. What is $\binom{n}{m}$?  
The number of different ways to order m successes in n trials  

4. How can you interpret $Î¸^m(1-Î¸)^{n-m}$?  
The probability of seeing m successes and (n-m) failures in a particular ordering  

5. What is $\sum_{k=i}^n \binom{n}{k}Î¸^k(1-Î¸)^{n-k}$?  
Cumulative likelihood of seeing i or more successes from n Bernoulli trials  

6. Using PMF, compute probability of at least 2 heads in 4 coin tosses given fair coin.  
$\sum_{k=i}^n \binom{n}{k}Î¸^k(1-Î¸)^{n-k}$  
i = 2, n = 4  
$\binom{4}{2}(0.5)^2(1-0.5)^{4-2} + \binom{4}{3}(0.5)^3(1-0.5)^{4-3} + \binom{4}{4}(0.5)^4(1-0.5)^{4-4}$  
= 11/16 or 0.6875  
Equivalent R code: pbinom(2,4,0.5)
